## Valid Python Notebooks

```python
@calculate_auto(CONFIG)
def valid_python_notebooks(notebooks, total_notebooks, prefix=""):
    return relative_var(
        prefix + "valid_python_notebooks",
        len(
            notebooks[
                (notebooks["language"] == "python")
                & (notebooks["language_version"] != "unknown")
                & ~(
                    (notebooks["kernel"] == "no-kernel")
                    & (notebooks["nbformat"] == "0")
                )
                & (notebooks["total_cells"] != 0)
                & (np.bitwise_and(notebooks["processed"], 16) == 0)
            ]
        ),
        total_notebooks,
    )
```

## Invalid Python Notebooks

```python
@calculate_auto(CONFIG)
def invalid_python_notebooks(notebooks, total_notebooks, prefix=""):
    return relative_var(
        prefix + "invalid_python_notebooks",
        len(
            notebooks[
                ~(
                    (notebooks["language"] == "python")
                    & (notebooks["language_version"] != "unknown")
                    & ~(
                        (notebooks["kernel"] == "no-kernel")
                        & (notebooks["nbformat"] == "0")
                    )
                    & (notebooks["total_cells"] != 0)
                    & (np.bitwise_and(notebooks["processed"], 16) == 0)
                )
            ]
        ),
        total_notebooks,
    )
```

## Python Version

```python
@calculate_auto(CONFIG, save_to="python_notebooks", result=None)
def python_version(notebooks, total_notebooks, prefix=""):
    notebooks.loc[:, "minor_version"] = notebooks["language_version"].str[:3]
    notebooks.loc[:, "major_version"] = notebooks["language_version"].str[:1]
    python_notebooks = notebooks[notebooks.language == "python"]
    # Removing inconsistencies
    python_notebooks = python_notebooks[
        ~python_notebooks["language_version"].isin({"3", "1", "ES2015"})
    ]
    return python_notebooks
```

# Execution Count

```python
@calculate_auto(CONFIG, save_to="without_execution_numbers", result=None)
def execution_count_without(python_notebooks, total_python_notebooks, prefix=""):
    without_execution_numbers = python_notebooks[
        python_notebooks.max_execution_count < 0
    ]
    n_without_execution_numbers = len(without_execution_numbers)
    print(
        relative_var(
            prefix + "non_executed_python_notebooks_relative_to_python_notebooks",
            n_without_execution_numbers,
            total_python_notebooks,
        )
    )
    return without_execution_numbers
```

```python
@calculate_auto(CONFIG, save_to="with_execution_numbers", result=None)
def execution_count_with(python_notebooks, total_python_notebooks, prefix=""):
    with_execution_numbers = python_notebooks[
        python_notebooks.max_execution_count >= -0
    ]
    n_with_execution_numbers = len(with_execution_numbers)
    print(
        relative_var(
            prefix + "executed_python_notebooks_relative_to_python_notebooks",
            n_with_execution_numbers,
            total_python_notebooks,
        )
    )
    return with_execution_numbers
```

#### Max execution count = number of non-empty code cells

```python
@calculate_auto(CONFIG)
def equal(diff, with_execution_numbers, prefix=""):
    return "{0:.2%}".format(len(diff[diff == 0]) / len(with_execution_numbers))
```

#### More code cells than execution numbers

```python
@calculate_auto(CONFIG)
def more(diff, with_execution_numbers, prefix=""):
    return "{0:.2%}".format(len(diff[diff < 0]) / len(with_execution_numbers))
```

#### Less code cells than execution numbers

```python
@calculate_auto(CONFIG)
def less(diff, with_execution_numbers, prefix=""):
    return "{0:.2%}".format(len(diff[diff > 0]) / len(with_execution_numbers))
```

```python
allowed = set(ascii_letters + digits + ".-_")
```

```python
@calculate_auto(CONFIG, save_to="invalid_names", result=None)
def invalid(notebook_name, total_notebooks, prefix=""):
    invalid_names = notebook_name[notebook_name.apply(lambda x: bool(set(x) - allowed))]
    total_invalid = len(invalid_names)
    print(
        "{} invalid names".format(
            relative_var(
                prefix + "not_recommended_names", total_invalid, total_notebooks
            )
        )
    )
    return invalid_names
```

```python
@calculate_auto(CONFIG)
def disal(invalid_names, total_notebooks, prefix=""):
    disallowed = set('<>:"/\\|?*')
    disallowed_names = invalid_names[
        invalid_names.apply(lambda x: bool(disallowed & set(x)))
    ]
    total_disallowed = len(disallowed_names)
    return "{} disallowed in windows".format(
        relative_var(prefix + "invalid_windows", total_disallowed, total_notebooks)
    )
```
